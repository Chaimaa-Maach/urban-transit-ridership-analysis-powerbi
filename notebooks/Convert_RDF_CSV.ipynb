# **Conversion des fichiers RDF**

### Chicago (RDF) : Transformation du format RDF vers CSV via un script Python.

```python
import pandas as pd
from rdflib import Graph,Namespace
import os
```

```python
# Dossiers
input_folder = "../data/raw/rdf-files/"
output_file = "../data/raw/chicago_ridership_all_routes.csv"

# Namespace
DS = Namespace("https://data.cityofchicago.org/resource/jyb9-n7fm/")

all_data = []

# Parcourir tous les fichiers RDF
for file_name in os.listdir(input_folder):
    if file_name.endswith(".rdf"):
        file_path = os.path.join(input_folder, file_name)
        print(f"Lecture de {file_name}...")

        # Charger le RDF
        g = Graph()
        g.parse(file_path, format="xml")

        # Extraire les données
        for obs in g.subjects(predicate=DS.route):
            route = g.value(obs, DS.route)
            date = g.value(obs, DS.date)
            daytype = g.value(obs, DS.daytype)
            rides = g.value(obs, DS.rides)

            if route and date and daytype and rides:
                all_data.append({
                    "route": str(route),
                    "date": str(date)[:10],  # YYYY-MM-DD
                    "day_type": str(daytype),
                    "rides": int(rides)
                })

# Créer le DataFrame global
df_all = pd.DataFrame(all_data)

# Sauvegarder en CSV unique
df_all.to_csv(output_file, index=False)
print(f"Tous les fichiers RDF ont été fusionnés en {output_file} avec {len(df_all)} lignes.")
```
